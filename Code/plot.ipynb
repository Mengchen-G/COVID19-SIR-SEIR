# Get the file

```
root_dir = '/content/drive/My Drive/Network Analysis and modeling/'    # add shortcut to drive, change root_dir to your drive 
base_dir = root_dir + 'Network Project/'
file = base_dir + 'Time.csv'
age_file = base_dir + 'TimeAge.csv'
gender_file = base_dir + 'TimeGender.csv'
province_file = base_dir + 'TimeProvince.csv'
```

# Import

```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# Read file

```
time_df = pd.read_csv(file, skiprows=0)
age_time_df = pd.read_csv(age_file, skiprows=0)
gender_time_df = pd.read_csv(gender_file, skiprows=0)
province_time_df = pd.read_csv(province_file, skiprows=0)
```

## plot time vs # infected cases

```
plt.figure(figsize = (5,5))
plt.plot(time_df['date'], time_df['confirmed'], label='confirmed')
plt.plot(time_df['date'], time_df['released'], label='released')
plt.plot(time_df['date'], time_df['deceased'], label='deceased')
plt.legend()
plt.xlabel('Time')
plt.ylabel('# infected cases')
```

## Plot time vs # infected cases affected by age

```
age_time_df = age_time_df.sort_values(['age', 'date'], ascending=[True, True])
```

```
plt.figure(figsize = (5,5))
for i in range(0, 1089, 121):
  plt.plot(age_time_df['date'][i:121 + i], age_time_df['confirmed'][i:121 + i], label=age_time_df['age'][i])
plt.legend()
plt.xlabel('Time')
plt.ylabel('# infected cases')
```

## plot time vs # infected cases affected by gender

```
gender_time_df = gender_time_df.sort_values(['sex', 'date'], ascending=[True, True])
```

```
plt.figure(figsize = (5,5))
for i in range(0, 242, 121):
  plt.plot(gender_time_df['date'][i:121 + i], gender_time_df['confirmed'][i:121 + i], label=gender_time_df['sex'][i])
plt.legend()
plt.xlabel('Time')
plt.ylabel('# infected cases')
```

## plot time vs # infected cases affected by province

```
province_time_df = province_time_df.sort_values(['province', 'date'], ascending=[True, True])
plt.figure(figsize = (8,10))
for i in range(0, 2771, 163):
  plt.plot(province_time_df['date'][i:163 + i], province_time_df['confirmed'][i:163 + i], label=province_time_df['province'][i])
plt.legend()
plt.xlabel('Time')
plt.ylabel('# infected cases')
```
